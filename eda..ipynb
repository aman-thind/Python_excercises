```python
def segments(l, n):
    return [l[i:i + n] for i in range(0, len(l), n)]
def boxplot(data, cols, ncols):
    for lst in segments(cols, ncols):
        fig, axes = plt.subplots(nrows=1, ncols=ncols, figsize=(10, 3), dpi=100)
        for idx in range(0, len(lst)):
            attr = lst[idx]
            data = pd.concat([data['R'], data[attr]], axis=1)
            sns.boxplot(x=attr, y='R', data=data, ax=axes[idx])
        plt.tight_layout()

boxplot(data, df, 2)
plt.show()
```
