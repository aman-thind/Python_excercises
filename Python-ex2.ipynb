# Q2

```python
def Min():
    arr = int(input('Enter the size of 1D Array:')) 
    x = []
    y = []
    z = []

    for i in range(0, arr):
        print('Enter the element for index', i, 'of first list:')
        x.append(float(input()))
  
    for i in range(0, arr):
        print('Enter the element for index', i, 'of second list:')
        y.append(float(input()))

    for i in range(0, arr):
        print('Enter the element for index', i, 'of third list:')
        z.append(float(input()))

    a = min(x)
    b = min(y)
    c = min(z)
    result = [a, b, c]
    result.sort()

    return np.array(result)
```

```python
#3c

# Mean of columns
mean_cols = np.array(np.mean(data, axis = 0))
print('Mean of each column:', mean_cols)

# Sum of columns
sum_cols = np.array(np.sum(data, axis = 0))
print('Sum of each column:', sum_cols)

# Max of columns
max_cols = np.array(np.max(data, axis = 0))
print('Max of each column:', max_cols)

# Min of columns
min_cols = np.array(np.min(data, axis = 0))
print('Min of each column:', min_cols)

# Mean of rows
mean_rows = np.array(np.mean(data, axis = 1))
print('Mean of each row:', mean_rows)

# Sum of each rows
sum_rows = np.array(np.sum(data, axis = 1))
print('Sum of each row:', sum_rows)

# Max of rows
max_rows = np.array(np.max(data, axis = 1))
print('Max of each row:', max_rows)

# Min of rows
min_rows = np.array(np.min(data, axis = 1))
print('Min of each row:', min_rows)
```
